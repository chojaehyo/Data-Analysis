### 거리계산 (파이썬 vs 넘파이)

```python
from math import sqrt

points = [[1, 1], [3, 2], [5, 7]]
x = [2, 3]

result = []

for point in points:
    result.append(sqrt((point[0]-x[0]) ** 2 + (point[1] - x[1]) ** 2))

result
```

```python
import numpy as np

points_arr = np.array(points)
x_arr = np.array(x)

def distMeasure(points_arr, x_arr):
    return np.sqrt(np.sum(((points_arr - x_arr) ** 2), axis=1))

distMeasure(points_arr, x_arr)
```

### 계단 오르내리기

#### - 파이썬에서 randint는 마지막 값이 포함됨, numpy에서 randint는 마지막값 포함되지 않음

#### - 파이썬

```python
import random

position = 0
steps =1000
walk = [position]

for i in range(steps):
    step = 1 if random.randint(0, 1) else -1
    position += step
    walk.append(position)
```

```python
import matplotlib.pyplot as plt

plt.plot(walk[:100])
```

#### - numpy

```python
arr = np.random.randint(0, 2, 1000)
arr = np.where(arr == 1, 1, -1)

walk = np.cumsum(arr)
print(walk[-1])
```

#### - 계단을 오르내린 위치의 최솟값과 최댓값 구하기

```python
print(walk.max(), walk.min())
```
